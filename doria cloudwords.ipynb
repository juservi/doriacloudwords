```python
from bs4 import BeautifulSoup
from pytube import YouTube

lista_atual = []
 
def coleta(link):
    yt = YouTube(link)
    caption = yt.captions.get_by_language_code('pt')
    arquivo = caption.xml_captions
    xml_soup = BeautifulSoup(arquivo, 'xml')
    linhas = xml_soup.find_all('text')

#print(linhas)
 
#lista_skaf_01 = list()

    palavras = list()

    for lin in linhas:
    
        txt = lin.get_text()
    
    #print(txt)
    
        sep_palavras = str(txt).split(' ')
        for sp in sep_palavras:
            palavras.append(sp)
            
    #print(palavras)
    
    for fim in palavras:
        lista_atual.append(fim)
    print(lista_atual)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('portuguese')) 
print(stop_words)
```

```python
lista_atual_limpa=[]
for i in lista_atual:
    if i not in stop_words:
        lista_atual_limpa.append(i)
print(lista_atual_limpa)
print(len(lista_atual_limpa))
```

```python
lista_atual_limpa_02 = []
for y in lista_atual_limpa:
    if "color" not in y:
        lista_atual_limpa_02.append(y)
```
